```python
def computer(input_int, phase_int):
    i = 0
    ans = command_string.copy()
    phase = True
 #   ans[1] = noun
 #   ans[2] = verb
    while i < len(command_string):
        instruction = str(ans[i]).zfill(5)
        operator = (int(str(instruction)[-2:]))
        mode1 = int((str(instruction)[-3:-2]))
        mode2 = int((str(instruction)[-4:-3]))
        mode3 = int((str(instruction)[-5:-4]))
#         print(ans[i],ans[i+1],ans[i+2],ans[i+3])
#         print(i)
#         print(operator)
#         print(mode1)
#         print(mode2)
        if(operator == 1):
            if(mode1 == 0):
                x1 = ans[ans[i+1]]
            else:
                x1 = ans[i+1]
            if(mode2 == 0):
                x2 = ans[ans[i+2]]
            else:
                x2 = ans[i+2]
#             print(x1, x2, x1*x2)
#             print(ans[ans[i+3]])
            ans[ans[i+3]] = x1 + x2
            i = i + 4
        if(operator == 2):
            if(mode1 == 0):
                x1 = ans[ans[i+1]]
            else:
                x1 = ans[i+1]
            if(mode2 == 0):
                x2 = ans[ans[i+2]]
            else:
                x2 = ans[i+2]
            ans[ans[i+3]] = x1 * x2
            i = i + 4
        if(operator == 3):
            i1 = phase_int if phase else input_int
            if(mode1 == 0):
                ans[ans[i+1]] = i1
            else:
                ans[i+1] = i1
            phase = False
            i = i + 2
        if(operator == 4):
            if(mode1 == 0):
                return([ans[ans[i+1]],4])
            else:
                return([ans[i+1],4])
            i = i + 2
        if(operator == 5):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(param1 != 0):
                i = param2
            else:
                i = i + 3
        if(operator == 6):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(param1 == 0):
                i = param2
            else:
                i = i + 3
        if(operator == 7):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
#            if(mode3 == 0):
#                 param3 = ans[ans[i+3]]
#             else:
            param3 = ans[i+3]
            if(param1 < param2):
                ans[param3] = 1
            else:
                ans[param3] = 0
            i = i + 4
        if(operator == 8):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
#             if(mode3 == 0):
#                 param3 = ans[ans[i+3]]
#             else:
            param3 = ans[i+3]
#            print(i, ans[i+1], ans[ans[i+1]], ans[ans[i+2]], param1, param2, param3)
            if(param1 == param2):
                ans[param3] = 1
            else:
                ans[param3] = 0
            i = i + 4
        if(operator == 99):
            return(-1, 99)
    #return(ans)
```

```python
def iterator(phase_list):
    input_int = 0
    i = 0
    out_operator = 0
    while out_operator != 99:
        out_operator = computer(input_int, phase_list[i])[1]
        print(input_int, phase_list[i], out_operator, i)
        if(out_operator != 99):
            input_int = computer(input_int, phase_list[i])[0]
        i += 1
        if(i > 4):
            i = 0
    return(input_int)
```

```python
outputs = []
import itertools
phase_lists = itertools.permutations([5,6,7,8,9])
for phase_list in phase_lists:
    outputs.append(iterator(phase_list))
print(max(outputs))
```
