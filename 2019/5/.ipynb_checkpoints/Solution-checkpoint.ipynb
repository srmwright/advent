```python
def computer(input_int):
    i = 0
    ans = command_string.copy()
 #   ans[1] = noun
 #   ans[2] = verb
    while i < len(command_string):
        instruction = str(ans[i]).zfill(5)
        operator = (int(str(instruction)[-2:]))
        mode1 = int((str(instruction)[-3:-2]))
        mode2 = int((str(instruction)[-4:-3]))
        mode3 = int((str(instruction)[-5:-4]))
        print(ans[i],ans[i+1],ans[i+2],ans[i+3])
        print(i)
        print(operator)
        print(mode1)
        print(mode2)
        if(operator == 1):
            if(mode1 == 0):
                x1 = ans[ans[i+1]]
            else:
                x1 = ans[i+1]
            if(mode2 == 0):
                x2 = ans[ans[i+2]]
            else:
                x2 = ans[i+2]
#             print(x1, x2, x1*x2)
#             print(ans[ans[i+3]])
            ans[ans[i+3]] = x1 + x2
            i = i + 4
        if(operator == 2):
            if(mode1 == 0):
                x1 = ans[ans[i+1]]
            else:
                x1 = ans[i+1]
            if(mode2 == 0):
                x2 = ans[ans[i+2]]
            else:
                x2 = ans[i+2]
            ans[ans[i+3]] = x1 * x2
            i = i + 4
        if(operator == 3):
            if(mode1 == 0):
                ans[ans[i+1]] = input_int
            else:
                ans[i+1] = input_int
            i = i + 2
        if(operator == 4):
            if(mode1 == 0):
                print(ans[ans[i+1]])
            else:
                print(ans[i+1])
            i = i + 2
        if(operator == 5):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(param1 != 0):
                i = param2
        if(operator == 6):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(param1 == 0):
                i = param2
        if(operator == 7):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(mode3 == 0):
                param3 = ans[ans[i+3]]
            else:
                param3 = ans[i+3]
            if(param1 < param2):
                ans[param3] = 1
            else:
                ans[param3] = 0
            i = i + 4
        if(operator == 8):
            if(mode1 == 0):
                param1 = ans[ans[i+1]]
            else:
                param1 = ans[i+1]
            if(mode2 == 0):
                param2 = ans[ans[i+2]]
            else:
                param2 = ans[i+2]
            if(mode3 == 0):
                param3 = ans[ans[i+3]]
            else:
                param3 = ans[i+3]
            if(param1 == param2):
                ans[param3] = 1
            else:
                ans[param3] = 0
            i = i + 4
        if(operator == 99):
            break
    return(ans)
```
