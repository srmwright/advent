### Part 1

```python
import copy
```

```python
### Input file into lines
with open('input.txt') as f:
    input = f.read()

lines = input.strip().split('\n')
```

```python
for i in range(0,len(lines)):
    lines[i] = list(lines[i])
```

```python
def prox(i,j,lines):
    adj = 0
    for x in range(-1,2):
        for y in range(-1,2):
            i0 = i + x
            j0 = j + y
            if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]):
                if lines[i0][j0] == '#':
                    adj += 1
    #print(i,j,adj)
    return adj
```

```python
lines2 = copy.deepcopy(lines)
```

```python
it = 0
while it >= 0:
    it += 1
    lines = copy.deepcopy(lines2)
    for i in range(0,len(lines)):
        for j in range(0, len(lines[0])):
            if lines[i][j] == 'L':
                adj = prox(i,j,lines)
                if adj == 0:
                    lines2[i][j] = '#'
            if lines[i][j] == '#':
                adj = prox(i,j,lines)
                if adj >= 5:
                    lines2[i][j] = 'L'
            if lines[i][j] == '.':
                lines2[i][j] = '.'
#     print(lines)
#     print(lines2)
    if lines == lines2:
        break
print(it)
occ = 0
for i in range(0,len(lines)):
        for j in range(0, len(lines[0])):
            if lines[i][j] == '#':
                occ +=1
print(occ)
```

### Part 2

```python
## THE FIRST SEAT THEY SEE IN EACH CARDINAL DIRECTION
def prox2(i,j,lines):
    adj = 0
    a = [0]*8
    nw = '.'
    n = '.'
    ne = '.'
    e = '.'
    se = '.'
    s = '.'
    sw = '.'
    w = '.'
    dist = [0]*8
    for x in range(1,len(lines)):
        ## northeast
        i0 = i - x
        j0 = j + x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and ne == '.':
            ne = lines[i0][j0]
        else:
            dist[0] = x - 1
            a[0] += 1
        ## north
        i0 = i - x
        j0 = j
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and n == '.':
            n = lines[i0][j0]
        else:
            dist[1] = x - 1
            a[1] += 1
        ## northwest
        i0 = i - x
        j0 = j - x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and nw == '.':
            nw = lines[i0][j0]
        else:
            dist[2] = x - 1
            a[2] += 1
        ## west
        i0 = i
        j0 = j - x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and w == '.':
            w = lines[i0][j0]
        else:
            dist[3] = x - 1
            a[3] += 1
        ## southwest
        i0 = i + x
        j0 = j - x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and sw == '.':
            sw = lines[i0][j0]
        else:
            dist[4] = x - 1
            a[4] += 1
        ## south
        i0 = i + x
        j0 = j
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and s == '.':
            s = lines[i0][j0]
        else:
            dist[5] = x - 1
            a[5] += 1
        ## southeast
        i0 = i + x
        j0 = j + x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and se == '.':
            se = lines[i0][j0]
        else:
            dist[6] = x - 1
            a[6] += 1
        ## east
        i0 = i
        j0 = j + x
        if i0 >= 0 and j0 >= 0 and i0 < len(lines) and j0 < len(lines[0]) and e == '.':
            e = lines[i0][j0]
        else:
            dist[7] = x - 1
            a[7] += 1
        if all(aa > 0 for aa in a):
            break
    adjs = [ne,n,nw,w,sw,s,se,e]
#     if i == 22 and j == 0:
#         print(adjs)
#         print(dist)
    #print(i,j,adjs,sum(i=="#" for i in adjs))
    #print(i,j,adj)
    return sum(i=="#" for i in adjs)
```

```python
it = 0
while it >= 0:
    it += 1
    lines = copy.deepcopy(lines2)
    b = 0
    for i in range(0,len(lines)):
        for j in range(0, len(lines[0])):
            if lines[i][j] == 'L':
                adj = prox2(i,j,lines)
                if adj == 0:
                    lines2[i][j] = '#'
            if lines[i][j] == '#':
                adj = prox2(i,j,lines)
                if adj >= 5:
                    lines2[i][j] = 'L'
            if lines[i][j] == '.':
                lines2[i][j] = '.'
            if lines2[i][j] == '#':
                b += 1
    #print(lines)
    #print(lines2)
    print(it, b)
    if lines == lines2:
        #print(lines)
        break
    if it == 100:
        break
print(it)
occ = 0
for i in range(0,len(lines)):
        for j in range(0, len(lines[0])):
            if lines[i][j] == '#':
                occ +=1
print(occ)
```
