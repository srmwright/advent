### Part 1

```python
import re
import copy
```

```python
### Input file into lines
with open('input.txt') as f:
    input = f.read()

lines = input.strip().split('\n')
```

```python
poses = []
for i in lines:
    instruction = re.findall("([a-z\[\]0-9]+) =",i)[0]
    mask = ''
    if instruction == 'mask':
        mask = re.findall("= ([X0-9]+)",i)[0]
    else:
        poses.append(int(re.findall("[0-9]+",i)[0]))
        mask = format(int(re.findall("= ([0-9]+)",i)[0]),'036b')
```

```python
mem = ['']*(max(poses)+1)
```

```python
mask = ''
for i in lines:
    instruction = re.findall("([a-z\[\]0-9]+) =",i)[0]
    if instruction == 'mask':
        mask = re.findall("= ([X0-9]+)",i)[0]
    else:
        pos = int(re.findall("[0-9]+",i)[0])
        binstring = format(int(re.findall("= ([0-9]+)",i)[0]),'036b')
        bs = list(binstring)
        for j in range(0,len(mask)):
            if mask[j] != 'X':
                bs[j] = mask[j]
        mem[pos] = ''.join(bs)
```

```python
new_mem = []
for i in range(0,len(mem)):
    if mem[i] == '':
        new_mem.append(0)
    else:
        new_mem.append(int(mem[i],2))
```

```python
sum(new_mem)
```

```python
ini_poses = []
for i in lines:
    instruction = re.findall("([a-z\[\]0-9]+) =",i)[0]
    mask = ''
    if instruction == 'mask':
        mask = re.findall("= ([X0-9]+)",i)[0]
    else:
        ini_poses.append(int(re.findall("[0-9]+",i)[0]))
        mask = format(int(re.findall("= ([0-9]+)",i)[0]),'036b')
```

```python
#mem = ['']*(100000*max(ini_poses)+1)
# can't do this, too much memory, will have to run linked key-value pairs...
mem_poses = ini_poses
mem_values = [0]*len(ini_poses)
mask = ''
for i in lines:
    print(i)
    instruction = re.findall("([a-z\[\]0-9]+) =",i)[0]
    if instruction == 'mask':
        mask = re.findall("= ([X0-9]+)",i)[0]
    else:  
        pos = format(int(re.findall("[0-9]+",i)[0]),'036b')
        value = int(re.findall("= ([0-9]+)",i)[0])
        p = list(pos)
        for j in range(0,len(mask)):
            if mask[j] == '1':
                p[j] = '1'
        poses = [p]
        for j in range(0,len(mask)):
            if mask[j] == 'X':
                for k in range(0,len(poses)):
                    for l in range(0,2):
                        p1 = copy.deepcopy(poses[k])
                        p1[j] = str(l)
                        poses.append(p1)
        pretty_poses = [int(''.join(p),2) for p in poses]
        #print(set(pretty_poses))
        for p in set(pretty_poses):
            if p in mem_poses:
                mem_values[mem_poses.index(p)] = value
            else:
                mem_poses.append(p)
                mem_values.append(value)
      #  mem[''.join(p)] = value
```
